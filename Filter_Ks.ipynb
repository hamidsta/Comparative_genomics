```python
df = pd.read_csv('Pair_Ks', sep='\t')
df.head(10)
```

```python
sd = df['Ks'].std()
```

```python
sd
```

```python
for index, row in df.iterrows():
    if row['Ks'] < sd:
        df = df.drop(index)
```

```python
df["pairs"] = df[["Protein1", "Protein2"]].apply("-".join, axis=1)
```

```python
df1 = df[['pairs','Ks']]
df1
```

```python
df1.to_csv('Ks_new_treshold.csv',index=False)
```
